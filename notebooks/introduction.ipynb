```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
```

Suponiendo que tenemos una moneda justa, asumimos que ambos únicos resultados de la moneda son igualmente probables, por lo que su probabilidad es $p=\frac{1}{2}$. Para distinguir cada volado utilizaremos la siguiente notación: $$X_n=\begin{cases}1&\text{n-ésimo volado es Águila}\\\\0&\text{n-ésimo volado es Sol}\end{cases}$$ Para simular estos volados crearemos una matriz ceros y unos de tamaño $m\times n$ donde cada elemento representa un resultado de un volado. De esta manera, cada renglón representa un experimento de $n$ volados.

```python
p = 1/2; # probabilidad de águila
n = 50; # número de volados
m = 5; # número de experimentos
y=np.random.binomial(1,p,size=(m,n)) # matriz de m x n lanzamientos de moneda
```

```python
y
```

Creamos gráficas para visualizar los resultados de cada experimento:

```python
x0 = np.arange(1,n+1,1); # para representar cada volado
for i in np.arange(0,m,1): #
    y0 = y[i,:]
    plt.figure(i)
    sns.scatterplot(x=x0,y=y0).set_yticks(range(2))
    plt.title(f"Experimento {i}")
    plt.xlabel("# de volado")
    plt.ylabel("Águila o Sol")
```

De estas gráficas no se alcanza a distinguir algun patrón entre experimentos, sin embargo, la suma acumulativa del número de águilas conseguidas en un experimento (50 volados) nos da una mayor pista, siendo esta la siguiente expresión: $$\begin{array}{cc}\displaystyle S_k=\sum_{i=1}^kX_i&k=1,\dots,n.\end{array}$$

```python
Y = np.cumsum(y,axis=1) # Número acumulado de Águilas
adjY = Y.conj().T
sns.lineplot(adjY)
plt.legend(title="Experimento", bbox_to_anchor=(1, 1))
plt.xlabel("Volados")
plt.ylabel("Águilas acumuladas")
```

```python
Y
```

Esta gráfica nos ayuda a distinguir las fluctuaciones del número de águilas obtenidas en cada experimento, donde mantienen una tendencia lineal creciente con patrones muy similares en sus fluctuaciones

```python
N = np.tile(np.arange(1,n+1,1),(m,1)) # matriz de números naturales
Y1 = Y/N # Acumulado dividido entre número de volados
sns.lineplot(np.conj(Y1).T).set_yticks(np.arange(0,1.1,0.5))
```

Al dividir el acumulado del número de águilas, encontramos la clara convergencia de cada experimento hacia el número de nuestra probabilidad.

```python
Y2=Y-p*N
sns.lineplot(np.conj(Y2).T)
```

```python
p = 1/2; # probabilidad de águila
n = 1000; # número de volados
m = 500; # número de experimentos
y=np.random.binomial(1,p,size=(m,n)) # matriz de m x n lanzamientos de moneda
Y = np.cumsum(y,axis=1)
N = np.tile(np.arange(1,n+1,1),(m,1))
Y2 = Y-p*N
sns.lineplot(np.conj(Y2).T)
plt.legend([],[],frameon=False)
```

```python
SN= np.tile(np.arange(1,n+1,1)**(0.5),(m,1))
Y3=Y2/SN
sns.lineplot(np.conj(Y3).T)
plt.legend([],[],frameon=False)
```
